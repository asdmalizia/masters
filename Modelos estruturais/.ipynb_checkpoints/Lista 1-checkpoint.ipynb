# Questão 2 - Simulações

```python
def state_space_model(Z, T, R, c, d, σ=None, N=120, fZ = None):
    
    m, r = T.shape[0], R.shape[1]
    if σ is None: σ = rng.random()*10
    Q = np.diag(rng.random(r)*10)
    
    ηt = rng.multivariate_normal(mean=np.zeros(r), cov=Q, size=N)
    ϵt = rng.normal(0, σ, N)
    ϵt = ϵt.reshape(1,-1)

    data = pd.DataFrame()
    for j in range(5):
        α1 = rng.random(m)*100
        αt = α1.copy()
        yt = np.zeros((1,N))
        
        for i in range(N):
            if fZ: Z = fZ(yt, i) 
            yt[:,i] = Z@αt + d + ϵt[:,i]
            αt = T@αt + c + R@ηt[i]
        data.loc[:,j] = yt[0]
    return data
```

## a) 
O vetor de estado é uma componente de nível local, que segue um passeio aleatório

```python
Z = np.array([[1]])
T = np.array([[1]])
R = np.array([[1]])
c, d = c, d = np.zeros(1), np.zeros(1)

data = state_space_model(Z, T, R, c, d)
```

```python
fig, ax = plt.subplots(5, figsize=(10,6))

for i in range(5):
    ax[i].plot(data.iloc[:,i])
ax[0].set_title('Simulações do modelo')
del data
plt.show()
```

## c) 
O modelo é um ARMA(2,1,1) não estacionário

```python
Z = np.array([[1,0,0]])

θ = 2
ϕ1, ϕ2 = 3, 5
T = np.array([[ϕ1,ϕ2,θ], [1,0,0], [0,0,0]])
R = np.array([[1,0,1]])
c, d = c, d = np.array([5, 0.3, 1]), np.zeros(1)
    
data = state_space_model(Z, T, R, c, d, σ=0)
data = data + data.shift(1, fill_value=5)
```

```python
fig, ax = plt.subplots(5, figsize=(10,6))

for i in range(5):
    ax[i].plot(data.iloc[:,i])
ax[0].set_title('Simulações do modelo')
del data
plt.show()
```

## d) Fazer
O vetor de estado é um componente de nível com slope determinístico.

```python
def fZ(y, t):
    if t==0:
        y0 = rng.random(yt.shape[0])*100
        return np.array([y0])
    return y[:,[t-1]]
```

```python
α = 0.7
β = 8
Z = np.array([[1]])
T = np.array([[α]])
R = np.array([[1]])
c, d = c, d = np.array([β]), np.zeros(1)
    
data = state_space_model(Z, T, R, c, d, fZ=fZ)
```

```python
fig, ax = plt.subplots(5, figsize=(10,6))

for i in range(5):
    ax[i].plot(data.iloc[:10,i])
ax[0].set_title('Simulações do modelo')
del data
plt.show()
```
