# Lista Prática 1

```python
import pandas as pd
import numpy as np
import requests
import json
import matplotlib.pyplot as plt
from datetime import datetime
from statsmodels.tsa.exponential_smoothing.ets import ETSModel
from statsmodels.graphics.tsaplots import plot_acf
```

# Questão 1

A venda de fertilizantes está diretamente relacionada com as condições financeiras das commodites NPK (Nitrogênio, Fósforo e Potássio), que são utilizadas como fertilizantes.  

Em um contexto de investimento em commodities por exemplo, prever a venda de fertilizantes é essencial para a tomada de decisão. Com a previsão das vendas em determinado horizonte, é possível entender a demanda futura pelas commodities NPK e usá-la na análise dos preços praticados pelo mercado.

## Download da Série

```python
url = 'http://www.ipeadata.gov.br/api/odata4/'
ticker = 'ANDA12_VFERTILIZ12'

# Dados
r = requests.get(url + "/Metadados('{}')/Valores".format(ticker))
data = json.loads(r.text)

series = pd.DataFrame(data['value'], columns=['VALDATA','VALVALOR']).rename(columns={'VALDATA':'date','VALVALOR':'value'}).set_index('date')
series.index = pd.to_datetime(series.index.map(lambda x: x.split('T')[0]))
series = series.div(1e6)

# Divisão entre período de teste e treino
series_train = series.iloc[:-12]
series_test = series.iloc[-12:]
```

# Questão 3

## item (i): Expressões ETS

- ETS(A,A,N): $ y_t = l_{t-1} + b_{t-1} + \epsilon_t $   
  $l_t = l_{t-1} + b_{t-1} + \alpha \epsilon_t $   
  $b_t = b_{t-1} + \beta \epsilon_t $   
  
- ETS(A,A,A):  $ y_t = l_{t-1} + b_{t-1} + s_{t-m} + \epsilon_t $   
  $l_t = l_{t-1} + b_{t-1} + \alpha \epsilon_t $   
  $b_t = b_{t-1} + \beta \epsilon_t $   
  $ s_t = s_{t-m} + \gamma \epsilon_t $

- ETS(A,A_d,A):  $ y_t = l_{t-1} + \phi b_{t-1} + s_{t-m} + \epsilon_t $   
  $l_t = l_{t-1} + \phi b_{t-1} + \alpha \epsilon_t $   
  $b_t = \phi b_{t-1} + \beta \epsilon_t $   
  $ s_t = s_{t-m} + \gamma \epsilon_t $

- ETS(M,A_d,M):  $ y_t = [(l_{t-1} + \phi b_{t-1})s_{t-m}](1+\epsilon_t) $   
  $l_t = (l_{t-1} + \phi b_{t-1}) (1+\alpha \epsilon_t) $   
  $b_t = \phi b_{t-1} + \beta(l_{t-1} + \phi b_{t-1}) \epsilon_t $   
  $ s_t = s_{t-m} (1+ \gamma \epsilon_t) $

- ETS(A,A,M):  $ y_t = (l_{t-1} + b_{t-1})s_{t-m} + \epsilon_t $   
  $l_t = l_{t-1} + b_{t-1} + \alpha \epsilon_t/s_{t-m} $   
  $b_t = b_{t-1} + \beta \epsilon_t/s_{t-m} $   
  $ s_t = s_{t-m} + \gamma \epsilon_t/(l_{t-1}+b_{t-1}) $

## item (ii): Expressões de previsão

- ETS(A,A,N): $\hat{y}_{t+h|t} = l_{t} + hb_{t} $   
  $l_t = \alpha y_t + (1-\alpha)(l_{t-1} + b_{t-1}) $   
  $b_t = \beta^*(l_t-l_{t-1}) + (1-\beta^*)b_{t-1} $   
  
- ETS(A,A,A): $ \hat{y}_{t+h|t} = l_{t} + hb_{t} + s_{t+h-m(k+1)} $   
  $l_t = \alpha (y_t-s_{t-m}) + (1-\alpha)(l_{t-1} + b_{t-1}) $   
  $b_t = \beta^*(l_t-l_{t-1}) + (1-\beta^*)b_{t-1} $   
  $ s_t = \gamma(y_t-l_{t-1}-b_{t-1}) + (1-\gamma)s_{t-m} $

  
- ETS(A,A_d,A): $ \hat{y}_{t+h|t} = l_{t} + (\phi +\phi^2+...+\phi^h)b_{t} + s_{t+h-m(k+1)} $   
  $l_t =  \alpha (y_t-s_{t-m}) + (1-\alpha)(l_{t-1} + \phi b_{t-1}) $   
  $b_t = \beta^*(l_t-l_{t-1}) + (1-\beta^*)\phi b_{t-1} $   
  $ s_t = \gamma(y_t-l_{t-1}-\phi b_{t-1}) + (1-\gamma)s_{t-m} $

- ETS(M,A_d,M): $ \hat{y}_{t+h|t} = [l_{t} + (\phi +\phi^2+...+\phi^h) b_{t}]s_{t+h-m(k+1)} $   
  $l_t = \alpha (y_t/s_{t-m}) + (1-\alpha)(l_{t-1} + \phi b_{t-1}) $   
  $b_t = \beta^*(l_t-l_{t-1}) + (1-\beta^*)\phi b_{t-1} $   
  $ s_t = \gamma(\frac{y_t} {l_{t-1}-\phi b_{t-1}}) + (1-\gamma)s_{t-m} $

- ETS(A,A,M): $ \hat{y}_{t+h|t} = (l_{t} + hb_{t})s_{t+h-m(k+1)} $   
  $l_t = \alpha (y_t/s_{t-m}) + (1-\alpha)(l_{t-1} + b_{t-1}) $   
  $b_t = \beta^*(l_t-l_{t-1}) + (1-\beta^*)b_{t-1} $   
  $ s_t = \gamma(\frac{y_t} {l_{t-1}- b_{t-1}}) + (1-\gamma)s_{t-m} $

## item (iii): Estimativa das constantes

Em todos os modelos, o parâmetro alpha foi estimado com valores próximos de 1, agregando à componente do level fortes variações estocásticas. Por outro lado, o beta foi estimado com valor perto de zero, o que implica em um slope quase constante. A componente sazonal também possui valores mais baixos do coeficiente gamma, tendo menos efeito na parcela estocástica.   
Já os dois coeficientes phi estimados estão, respectivamente, nos valores mínimo e máximo propostos no livro de Hyndman, possuindo um forte efeito de damp no primeiro caso e fraco no segundo.

```python
# definição dos modelos
df_params = pd.DataFrame(columns = ['model','alpha','beta','gamma','phi','k_params'])
model_params = [
    ['add','add',None,False], # ETS(A,A,N)
    ['add','add','add',False], # ETS(A,A,A)
    ['add','add','add',True], # ETS(A,Ad,A)
    ['mul','add','mul',True], # ETS(M,Ad,M)
    ['add','add','mul',False] # ETS(A,A,M)
]

# estimação dos cinco modelos
model_fited = []
for params in model_params:
    model = ETSModel(series_train['value'], 
                     error=params[0], trend=params[1], seasonal=params[2], 
                     damped_trend=params[3], seasonal_periods=12,
                    freq='M', dates=series_train.index.to_period('M'))
    fit = model.fit()
    model_fited.append(fit)
    df_params.loc[df_params.shape[0]] = [model.short_name, fit.smoothing_level, fit.smoothing_trend,
                                        fit.smoothing_seasonal if fit.has_seasonal else None, fit.damping_trend if fit.damped_trend else None, fit.k_params]

# resultados
df_params.round(4)
```

## item (iv): Métricas de aderência

De acordo com a tabela de resultados, o melhor modelo no período de treinamento é o ETS(A,A,A), pois possui o menor AICc.

```python
df_metrics = pd.DataFrame(columns=['model','AICc','RMSE','MAD','MAPE','SMape','MASE'])

# cálculo das métricas pra cada modelo estimado
for model in model_fited:
    mape = model.resid.abs().div(series_train['value']).mean()
    smape = model.resid.abs().div(
        series_train['value'].add(model.fittedvalues).div(2)
    ).mean()
    mase = model.mae/(
        series_train['value'].sub(series_train['value'].shift()).abs().mean()
    )
    
    df_metrics.loc[df_metrics.shape[0]] = [
        model.short_name, model.aicc, model.mse**0.5, model.mae, 
        100*mape, 100*smape, mase]                            

# resultados
df_metrics.round(4)
```

## item (v): FAC dos resíduos

Com exceção da FAC dos resíduos do ETS(A,A,N), todos os demais apresentam baixa ou insignificativa autocorrelação com os lags. Isso demonstra que esses modelos explicam bem a estrutura da série no período de treinamento, já que o resíduo não apresenta estrura de dependência.

No caso do modelo ETS(A,A,N), a sazonalidade não foi capturada, já que não há essa componente no modelo. Isso fica claro ao verificar no gráfico a forte autocorrelação dos resíduos no lag 12. 

Por fim, ambos modelos com sazonalidade multiplicativa, ETS(M,Ad,M) e ETS(A,A,M), apresentaram uma fraca autocorrelação no lag 12, o que indica que a combinação multiplicativa da componente de sazonalidade não é a melhor opção.

```python
lags=15
fig, ax = plt.subplots(5, figsize=(10, 15))

for i in range(len(model_fited)):
    model = model_fited[i]
    plot_acf(model.resid,lags=lags, ax=ax[i], title=model.short_name)
    ax[i].set_xticks(np.arange(lags+1))
    ax[i].set_xticklabels(np.arange(lags+1))

fig.suptitle('FAC dos Resíduos', size=16, x=0.12)
plt.tight_layout()
plt.show()
```

## item (vi): Previsões de teste

$\textbf{a)}$ De acordo com todas as métricas de adesão fora da amostra na tabela abaixo, o melhor modelo no período de teste é o ETS(A,Ad,A), que não coincide com o modelo escolhido no período de treinamento. Ainda assim eu escolheria o ETS(A,Ad,A), pois apesar de ter o AICc maior que o ETS(A,A,A), a diferença é pequena e ele possui métricas melhores fora da amostra.

```python
df_metrics_test = pd.DataFrame(columns=['model','RMSE','MAD','MAPE','SMape','MASE'])

# 5 fitted models
for model in model_fited:
    resid = series_test['value'].sub(model.forecast(12).values)
    mape = resid.abs().div(series_test['value']).mean()
    smape = resid.abs().div(
        series_test['value'].add(model.forecast(12).values).div(2)
    ).mean()
    mase = resid.abs().mean()/(
        series_train['value'].sub(series_train['value'].shift()).abs().mean() # naive calculado insample
    )
    
    df_metrics_test.loc[df_metrics_test.shape[0]] = [
        model.short_name, resid.pow(2).sum()**0.5, resid.abs().mean(), 
        100*mape, 100*smape, mase]

# naive predictions
predictions = {
    'naive1': [series_train['value'].iloc[-1]]*12, # ultima observação de treino da ST
    'naive2': series_train['value'].iloc[-12:].values # última observação de treino da ST para cada horizonte (12 meses)
}
for model, pred in predictions.items():
    resid = series_test['value'].sub(pred)
    mape = resid.abs().div(series_test['value']).mean()
    smape = resid.abs().div(
        series_test['value'].add(pred).div(2)
    ).mean()
    mase = resid.abs().mean()/(
        series_train['value'].sub(series_train['value'].shift()).abs().mean() # naive calculado insample
    )
    
    df_metrics_test.loc[df_metrics_test.shape[0]] = [
        model, resid.pow(2).sum()**0.5, resid.abs().mean(), 
        100*mape, 100*smape, mase]

df_metrics_test.round(4)
```

$\textbf{b)}$ Nos resultados abaixo, eu comparo as previsões do modelo calculadas pela expressão de previsão $ \hat{y}_{t+h|t} = l_{t} + (\phi +\phi^2+...+\phi^h)b_{t} + s_{t+h-m(k+1)} $ com os valores produzidos pelo programa, que são equivalentes.   
Para o cálculo da expressão, utilizei os componentes de level, trend, sazonalidade e damped estimados pelo modelo no último ano do período de treinamento. $l_t$ e $b_t$ são as últimas estimações do level e trend no preíodo de treinamento, referente a Junho/2023, enquanto a componente de sazonalidade é usada de acordo com o mês de previsão.

```python
y_h1 = (
    model_fited[2].level[-12:][-1] + #l_t
    model_fited[2].damping_trend * model_fited[2].slope[-12:][-1] + # \phi*b_t
    model_fited[2].season[-12:][-12] #s_{t-11}
)
y_h2 = (
    model_fited[2].level[-12:][-1] + #l_t
    model_fited[2].damping_trend * model_fited[2].slope[-12:][-1] + # \phi*b_t
    model_fited[2].damping_trend**2 * model_fited[2].slope[-12:][-1] + # \phi^2*b_t
    model_fited[2].season[-12:][-11] #s_{t-10}
)
y_h12 = (
    model_fited[2].level[-12:][-1] + #l_t
    sum([model_fited[2].damping_trend**x for x in range(1,13)]) * model_fited[2].slope[-12:][-1] + # (\phi + ... + \phi^12)*b_t
    model_fited[2].season[-12:][-1] #s_{t}
)
print('------- h=1 -------')
print('Manual: {} \nModelo: {}\n'.format(y_h1, model_fited[2].forecast(12)[0]))

print('------- h=2 -------')
print('Manual: {} \nModelo: {}\n'.format(y_h2, model_fited[2].forecast(12)[1]))

print('------- h=12 -------')
print('Manual: {} \nModelo: {}\n'.format(y_h12, model_fited[2].forecast(12)[-1]))
```

## item (vii): Melhor Modelo

$ \textbf{a)} $ Observando o gráfico dos componentes estimados do modelo, percebe-se o efeito da magnitude dos coeficientes de cada componente na sua variação estocástica. Enquanto o level possui forte efeito do coeficiente alpha na variança, a componente sazonal tem um efeito menor, variando menos nos meses.   
Além disso, é verificado um forte efeito do damped trend, que tende pra zero no começo da série. Esses resultados são coerentes, uma vez que a série escolhida apresenta forte sazonalidade e uma fraca tendência.

```python
# estimação do melhor modelo na série completa
best_model = ETSModel(series['value'], 
                     error='add', trend='add', seasonal='add', 
                     damped_trend=True, seasonal_periods=12,
                    freq='M', dates=series_train.index.to_period('M'))
best_fit = best_model.fit()

# plot do gráfico
fig, ax = plt.subplots(5, figsize=(10,12))
ax[0].set_ylabel('Observado')
ax[0].plot(series_train['value'])
ax[1].set_ylabel('Level')
ax[1].plot(best_fit.level)
ax[2].set_ylabel('Trend')
ax[2].plot(best_fit.slope)
ax[3].set_ylabel('Seasonal')
ax[3].plot(best_fit.season)
ax[4].set_ylabel('Resíduo')
ax[4].plot(best_fit.resid)

fig.suptitle('Componentes do Melhor Modelo')
fig.tight_layout()
plt.show()
```
